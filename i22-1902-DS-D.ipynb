```python
# YOUR CODE HERE
# Dictionary to store customers details

customers = {}

#Frozen Yogurt Menu 

yogurt_menu = {
    'size' : {
        'S' : 3.09,
        'M' : 3.89,
        'L' : 4.69
    },
    'container' : {
        'V' : 0,
        'U' : 0.1
    },
    'topping': {
        'G': 1.59, 
        'N': 1.89
    },
    'cookies': {
        'R': 1.49, 
        'C': 1.59
    },
    'size_choice': {
        'R': 0, 
        'L': 1.00
    },
    'Total' : 0,
    'Price' : 0.0
}

#Ice-Cream Menu 

ice_menu = {
    'size' : {
        'S' : 3.49,
        'M' : 4.49,
        'L' : 5.49
    },
    'container' : {
        'V' : 0,
        'U' : 0.1
    },
    'topping': {
        'G': 1.59, 
        'N': 1.89
    },
    'cookies': {
        'R': 1.49, 
        'C': 1.59
    },
    'size_choice': {
        'R': 0, 
        'L': 1.00
    },
    'Total' : 0,
    'Price' : 0.0
}

#Start of Inputs 

while True :
    print("\n*************************************")
    name = input("Enter your name").capitalize()

    #If old customer then Welcomeback
    
    if name in customers:
        print("Welcome back")

    #Total varible for calculation of total amount.

    total = 0.0

    #Asking Between 2 options

    while True :
        choice = input("Do you want Frozen Yogurt (Y), Ice Cream (I)? ").upper()
        if choice == 'Y':
            item = 'Frozen Yogurt'
            menu = yogurt_menu
            break
        elif choice == 'I':
            item = 'Ice Cream'
            menu = ice_menu
            break
        else:
            print("Invalid input. Please enter Y for Frozen Yogurt, I for Ice Cream")

    #Price variable for next inputs

    price = 0.0

    #Assigning prices from the dictionaries 

    while True:
        size = input("Enter Size : (S: Small, M: Medium, L: Large)").upper()
        if size in menu['size'] :
            price += menu['size'][size]
            break
        else :
            print("Invalid input.")

    while True:
        cont = input("Enter Container : (V: Cone, U: Cup)").upper()
        if cont in menu['container'] :
            price += menu['container'][cont]
            break
        else :
            print("Invalid input.")

    #Boolean of asking for toping that if they want or not

    topping_bool = False

    while True:
        topping_ch = input("Do you like to add topping : (Y: Yes, N: No)").upper()
        if topping_ch == 'Y' :
            topping_bool = True
            break
        elif topping_ch == 'N' :
            topping_bool = False
            break
        else :
            print("Invalid input.")

    #If customers need toping

    if topping_bool :
        while True :
            topping = input("Enter Topping : (G: Gummies, N: Nuts)").upper()
            if topping in menu['topping'] :
                price += menu['topping'][topping]
                break
            else :
                print("Invalid input.")

     #Boolean of asking for Cookies that if they want or not
    
    cookies_bool = False
    
    while True:
        cookies_ch = input("Do you like to add cookies : (Y: Yes, N: No)").upper()
        if cookies_ch == 'Y' :
            cookies_bool = True
            break
        elif cookies_ch == 'N' :
            cookies_bool = False
            break
        else :
            print("Invalid input.")

    #If customers need Cookies
    
    if cookies_bool :
        while True :
            cookies = input("Which Cookie do you like to add : (R: Raisin, C: Choclate)").upper()
            if cookies in menu['cookies'] :
                price += menu['cookies'][cookies]
                break
            else :
                print("Invalid input.")

    #Asking for size of Cookies 

    if cookies_bool :
        while True :
            size_choice = input("Which Cookie Size you want: (R: Regular, L : Large)").upper()
            if size_choice in menu['size_choice'] :
                price += menu['size_choice'][size_choice]
                break
            else :
                print("Invalid input.")

    #Asking for the days 
            
    while True:
        day_choice = input("What day of the week is this? (1: Monday, 2: Tuesday, 3: Wednesday, 4: Thursday, 5: Friday, 6: Saturday, 7: Sunday) ")
        if day_choice in ('1', '2', '3', '4', '5', '6', '7'):
            day_choice = int(day_choice)
            break
        else:
            print("Invalid input.")

    #Updating the total sales for yogurt and icexream in dictionary for report generation

    if choice == 'Y':
        yogurt_menu['Total'] += 1
        yogurt_menu['Price'] += price

    if choice == 'I':
        ice_menu['Total'] += 1
        ice_menu['Price'] += price

    #Printing the total of the order 

    print("Your Order Total is ", price)

    #Updating the customer info in records

    if name in customers :
        customers[name]['Total'] += price
        customers[name]['Orders'] += 1
    else :
        customers[name] = {
            'Total' : price,
            'Orders' : 1
        }

    #Asking for if they want to continue or return 

    continue_ch = input("Enter Q: quit ").upper()
    if continue_ch == 'Q' :
        break

#Printing all customers details

print("\n\n*************************************")
print("Customer Name\tOrder Total")
for idx, data in customers.items() :
    print("{:}\t\t{:.2f}".format(idx, data['Total']))


#Assigning variables for yogurt and Icrcream sale report

yogurt_tot = int(yogurt_menu['Total'])
yogurt_price = yogurt_menu['Price']
yogurt_avg = 0
if yogurt_tot != 0 :
    yogurt_avg = yogurt_price/yogurt_tot

ice_tot = int(ice_menu['Total'])
ice_price = ice_menu['Price']
ice_avg = 0
if ice_tot != 0 :
    ice_avg = ice_price/ice_tot

Grand_tot = int(yogurt_tot + ice_tot)
Grand_price = ice_price + yogurt_price
Grand_avg = ice_avg + yogurt_avg

#Printing the report 

print("\n\nItem\t\t#Order\tOrder Total\tAverage Per Order")
print("Yogurt\t\t{:}\t{:.2f}\t\t{:.2f}".format(yogurt_tot, yogurt_price, yogurt_avg))
print("IceCream\t{:}\t{:.2f}\t\t{:.2f}".format(ice_tot, ice_price, ice_avg))
print("Total\t\t{:}\t{:.2f}\t\t{:.2f}".format(Grand_tot, Grand_price, Grand_avg)) 
```

```python
# YOUR CODE HERE

#At first making the image pixel perfect 
#Image with (x<150,x<150,x<150) is considered as (0,0,0)

for i in range(height) :
    for j in range(width) :
        if(img_array[i][j][0] < 150 and img_array[i][j][1] < 150 and img_array[i][j][2] < 150) :
            img_array[i][j][0] = 0;
            img_array[i][j][1] = 0;
            img_array[i][j][2] = 0;
        else :
            img_array[i][j][0] = 255;
            img_array[i][j][1] = 255;
            img_array[i][j][2] = 255;

#Making and Printing the modified image

modified_img = Image.fromarray(img_array)
display(modified_img)

#Masking the white color with Yellow one

for i in range(height) :
    for j in range(width) :
        if(img_array[i][j][0] == 255) :
            img_array[i][j][0] = 255;
            img_array[i][j][1] = 255;
            img_array[i][j][2] = 0;

#Making and Printing the modified image

modified_img = Image.fromarray(img_array)
display(modified_img)  

#Masking the black color with white one

for i in range(height) :
    for j in range(width) :
        if(img_array[i][j][0] == 0) :
            img_array[i][j][0] = 255;
            img_array[i][j][1] = 255;
            img_array[i][j][2] = 255;

#Making and Printing the modified image

modified_img = Image.fromarray(img_array)
display(modified_img)
```

```python
# YOUR CODE HERE

#Library of Pandas
import pandas as pd

#Giving the file path to variable

file = pd.ExcelFile('TimeTable.xlsx')

#A list containing all data frames

all_DFs = []

#Keywords for matching of labs 

lab_keywords = ['C-Margala 1', 'C-Margala 3', 'C-Margala 4', 'C-Rawal 1', 'Rawal 3 (B-232)', 'C-Rawal 4', 'C-GPU Lab', 'A-Karakoram 1', 'A-Karakoram 2', 'A-Karakoram 3', 'A-Mehran 1', 'A-Mehran 2', 'B-Digital', 'A-CALL-1', 'A-CALL-2', 'A-CALL-3']

#For loop in order to look into all sheets

for sheet in file.sheet_names[1:]:

    #Read sheets one by one 
    
    df = pd.read_excel(file,sheet,header=4)

    #Drop columns containing all nulls
    
    df = df.dropna(axis=1,how='all')

    #List containing data 
    
    data = []

    #Prev time for assigning prev time to index is current time is empty
    
    prev_time = None

    #Variable for storing lab time
    
    lab_time = None

    #Iterating through each index in the excel sheet
    
    for i, row in enumerate(df.index):
        for j,col in enumerate(df.columns):

            #Assigning the index to cell
            
            cell = df.at[row,col]

            #For time if it is Unnamed then assign the prev col time to it else assign the 1st row for respective column to it
            
            if 'Unnamed' in col:
                time = prev_time
            else:
                time = df.columns[j]
                prev_time = time

            start_index = str(cell).find('(')
            end_index = str(cell).find(')')
            extracted_string = str(cell)[start_index + 1:end_index]

            #Creation of new row as Class = 1st col of respected row Day = SheetName and Course = Str(index)

            new_row = {
                'Class' : df.at[i, df.columns[0]],
                'Day' : sheet,
                'Course' : str(cell).split('(')[0],
            }

            if start_index != -1 and end_index != -1:
                new_row['Section'] = extracted_string
            else:
                new_row['Section'] = new_row['Course']

            #Now checking for time and type

            if new_row['Class'] in lab_keywords:
                new_row['Type'] = 'Lab'
                lab_time = df.iat[36, j] if j < len(df.columns) else ''
                new_row['Time'] = lab_time

            #Else it would be type = theory and time = time
            
            else:
                new_row['Type'] = 'Theory'
                new_row['Time'] = time 

            #Append the new rows into the data 
            
            data.append(new_row)

    #Making the data frame of the data 
    
    new_data = pd.DataFrame(data)

    #Remove all null rows
    
    new_data.dropna(inplace = True)

    #Remove all rows containing ''
    
    new_data = new_data[~(new_data == '').any(axis = 1)]

    #Remove all rows containing nan
    
    new_data = new_data[~(new_data == 'nan').any(axis = 1)]

    #Remove all rows containing time = room
    
    new_data = new_data[~(new_data['Time'] == 'Room')]

    #Remove all rows containing time = lab
    
    new_data = new_data[~(new_data['Time'] == 'Lab')]

    #Remove all rows containing class = lab

    new_data = new_data[~(new_data['Class'] == 'Lab')]

    #Reset the indexes in data frams
    
    new_data.reset_index(drop = True, inplace = True)

    #Append data frame to ALL DFS list and continue the loop till end of lists
    
    all_DFs.append(new_data)


#Now displaying and saving the dafa frames into csv file

for idx, df in enumerate(all_DFs):
    csv_filename = f'{file.sheet_names[idx]}.csv'
    df.to_csv(csv_filename, index=False)
    print(df)
    print('\n\n\n')
```
